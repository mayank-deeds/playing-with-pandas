```python
import pandas as pd
```

```python
 city_data = {
"City": ["New York City", "Paris", "Barcelona", "Rome"],
"Country": ["United States", "France", "Spain", "Italy"],
"Population": [8600000, 2141000, 5515000, 2873000]
}
cities = pd.DataFrame(city_data)
cities
```

```python
def fun(temp):
    temp+=1
    
    return temp
```

```python
# so this is the way to apply in place changes to the columns of a df without actually using the inplace thing 
cities.Population=cities.Population.apply(fun)
```
